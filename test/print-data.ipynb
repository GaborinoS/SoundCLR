```python
#print from the sound data 1-28808-A-43.wav the mel spectrogram and the mfcc    

import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import IPython.display as ipd
import random

#load the data
data, sampling_rate = librosa.load('1-28808-A-43.wav')

#plot the mel spectrogram
plt.figure(figsize=(9, 6))
librosa.display.specshow(librosa.power_to_db(librosa.feature.melspectrogram(data, sr=sampling_rate, n_mels=128, fmax=8000)), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')
plt.tight_layout()
plt.show()



```
